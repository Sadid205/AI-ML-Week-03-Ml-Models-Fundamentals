### Q1. Outlier Detection and Handling

You collected daily website traffic data:

```python
traffic = [120, 135, 140, 125, 138, 142, 130, 900]
```

1. Detect the outlier using the **IQR method**. You can either show calculations or explain the idea.
2. Suggest **one** appropriate way to handle this outlier and justify it in one line.
3. Give an example of a real life situation where this outlier should **not** be removed.

````python
# Optional: calculate IQR to detect the outlier
import numpy as np

traffic = np.array([120, 135, 140, 125, 138, 142, 130, 900])

Q1 = np.quantile(traffic,0.25)
Q3 = np.quantile(traffic,0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
no_outliers = traffic[(traffic  >= lower) & (traffic <= upper)]
print(no_outliers)
````

Use the code cell above if you want to quickly check quartiles and IQR. Explain your reasoning below.

**Your answer for Q1:**

1. The best method to detect outliers is using the Interquartile Range (IQR).
In my code above, I used np.quantile() to calculate the 25th percentile (Q1) and the 75th percentile (Q3) of the data.
Using these two values, I computed the lower bound and upper bound, and then used these bounds to detect and remove the outliers from the traffic array.

### Q3. Binning or Discretization

A continuous variable is given:

```python

```

1. Create **three bins** such as `young`, `middle`, `old` and assign each age to a bin.
2. State **one benefit** and **one drawback** of using binning in a machine learning model.

````python
# Optional: try binning using pandas
import pandas as pd

ages = [18, 20, 45, 67, 72, 23]
ages_series = pd.Series(ages, name="Age")
binning = pd.cut(
    ages_series,
    bins=[0,20,70,100],
    labels=["young","middle","old"]
)
print(binning)
````

**Your answer for Q3 (bins, benefit, drawback):**

### Q5. Concept of Regression and Line Fitting

You are predicting exam scores based on hours studied.

```python
hours = [1, 2, 3, 4, 5]
scores = [50, 55, 65, 70, 80]
```

1. In your own words, describe what **line fitting** means in linear regression.
2. What does the **slope** of the line represent in this context, in plain language?

````python
# Optional: fit a simple linear regression model
import numpy as np
from sklearn.linear_model import LinearRegression

hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
scores = np.array([50, 55, 65, 70, 80])

model = LinearRegression()
model.fit(hours, scores)
pred = model.predict(hours)
print("predictions",pred)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
````

**Your answer for Q5:**
Line fitting means finding the best possible straight line that represents the relationship between the input feature (like number of rooms, distance, age, etc.) and the target value you want to predict.
The line should be positioned in such a way that it is as close as possible to all the data points.


The slope tells us how much the prediction changes when the input changes.
In simple words:

A positive slope means: â€œAs the input increases, the output also increases.â€

A negative slope means: â€œAs the input increases, the output decreases.â€

The bigger the slope, the stronger the effect of the input on the output.

**Your answer for Q6:**
1. What does the cost function (like Mean Squared Error) measure?

The cost function measures the average amount of error between the modelâ€™s predictions and the true values.
Mean Squared Error (MSE) calculates the average squared difference, which gives more weight to larger errors.

2. What does gradient descent do to the cost function value step by step?

Gradient descent gradually reduces the cost function value.
At each step, it looks at the slope (gradient) of the cost function and updates the modelâ€™s parameters a little bit to reduce the error and improve the model.

3. Why is using a very large learning rate risky?

A very large learning rate can make the parameter updates too big, causing the cost function to increase instead of decrease, or making the modelâ€™s training unstable. As a result, the model may fail to learn properly or never reach the best solution.

### Q7. Regression Metrics Interpretation

A regression model produced the following metrics:

- R squared = 0.75
- MAE = 4.2
- RMSE = 7.6

1. Explain what each of these numbers tells you about the model.
2. Which metric **penalizes big errors more**, and why?


````python
# Optional: try computing these metrics on a toy example
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

true_vals = np.array([10, 12, 15, 20])
pred_vals = np.array([11, 13, 14, 19])

print("R squared:", r2_score(true_vals, pred_vals))
print("MAE:", mean_absolute_error(true_vals, pred_vals))
print("RMSE:", mean_squared_error(true_vals, pred_vals))
````

**Your answer for Q7:**
1. R squared = 0.75

R squared tells how much of the variation in the target variable the model explains.
Here, 0.75 means the model explains 75% of the variation in the data, which is quite good.

2. MAE = 4.2 (Mean Absolute Error)

MAE is the average absolute difference between predicted and actual values.
The modelâ€™s predictions are on average off by about 4.2 units.

3. RMSE = 7.6 (Root Mean Squared Error)

RMSE is the square root of the average squared errors, giving more weight to large errors.
Here, the average error is about 7.6, which is higher than MAE because RMSE penalizes larger errors more strongly.

### Q9. From Linear Regression to Logistic Regression

You want to predict whether a customer will buy a product, where `0` means No and `1` means Yes.

1. Why is **linear regression** not a good choice for this classification problem?
2. What role does the **sigmoid function** play in logistic regression?
3. If the sigmoid output is `0.81`, what is the predicted class when the decision threshold is `0.5`?

**Your answer for Q9:**
1. Why is linear regression not a good choice for classification?

Linear regression outputs continuous values that can be outside the range [0,1], but classification requires discrete classes (0 or 1). Therefore, linear regression cannot properly model class boundaries.

2. What role does the sigmoid function play in logistic regression?

The sigmoid function converts the input to a probability between 0 and 1, representing the likelihood of the data belonging to class 1, which helps in making classification decisions.

3. If the sigmoid output is 0.81 and the threshold is 0.5, what is the predicted class?

Since 0.81 > 0.5, the predicted class is 1 (Yes).
